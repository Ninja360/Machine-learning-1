```python
import numpy as np
```

```python
import pandas as pd
```

```python
df=pd.read_csv('CAMPUS.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
from sklearn.model_selection import train_test_split
```

```python
df['ssc_p'].fillna(df['ssc_p'].median(),inplace = True)
df['hsc_p'].fillna(df['hsc_p'].median(),inplace = True)
df['degree_p'].fillna(df['degree_p'].mean(),inplace = True)
```

```python
df['status'].replace(to_replace ="Placed", 
                 value =1,inplace=True) 
df['status'].replace(to_replace ="Not Placed", 
                 value =0,inplace=True) 
df['workex'].replace(to_replace ="Yes", 
                 value =1,inplace=True)
df['workex'].replace(to_replace ="No", 
                 value =0,inplace=True) 
df.head()
```

```python
drop=['sl_no','gender','ssc_b','hsc_b','hsc_s','degree_t','specialisation','salary']
```

```python
df.drop(drop,inplace=True,axis=1)
```

```python
df
```

```python
x=df.iloc[:,0:6]
```

```python
x.head()
```

```python
y=df.iloc[:,6]
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.5,random_state=999)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_train,y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators = 1000, random_state = 999)
```

```python
rf.fit(x_test, y_test)
```

```python
rf.score(x_test,y_test)
```
